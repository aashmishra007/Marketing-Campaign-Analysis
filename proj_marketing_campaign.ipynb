```python
# Removing leading and trailing spaces from column names
data.columns = data.columns.str.strip()
data.columns
```

```python
#Converting data type of 'Dt_Customer' column from object to datetime
data['Dt_Customer'] = pd.to_datetime(data['Dt_Customer'])
data['Dt_Customer'].dtype
```

<div style="text-align:left; color:yellow; font-size:20px;">Result:<br></div>
<div style="text-align:left; color:white; font-size:15px;">
<ol>

<li>Stored with nanosecond precision</li></ol></div>

```python
#Cleaning Income column by removing non-numeric characters and converting to numeric type
data['Income'] = pd.to_numeric(data['Income'].astype(str).str.replace('[^0-9.]', '', regex=True), errors='coerce')
```

```python
data['Income'].head()
```

<div style="text-align:left; color:Blue; font-size:25px;"><b>STEP 2</b></div>
Conduct missing value imputation, considering that customers with similar education and marital status tend to have comparable yearly incomes, on average.

```python
#Identifying rows with missing values in Income column
#data[data.isnull().any(axis=1)]
data[data['Income'].isnull()]
```

```python
group_mean_income = (data.groupby(['Education', 'Marital_Status'])['Income'].mean())
group_mean_income
```

```python
#filling missing values in Income column with mean income based on Education and Marital_Status
data['Income'] = data['Income'].fillna(data.groupby(['Education', 'Marital_Status'])['Income'].transform('mean'))
```

```python
#Verified that the missing value for ID 8996 has been filled with the group mean income for Phd and Married
data.loc[data['ID'] == 8996, 'Income']
```

<div style="text-align:left; color:Blue; font-size:25px;"><b>STEP 3(a)</b></div>
Create variables to represent the total number of children, age, and total spending.

```python
#Created variables to represent the total number of children, age, and total spending.
data['Total_Children'] = data['Kidhome'] + data['Teenhome']
data['Age'] = pd.Timestamp.today().year - data['Year_Birth']
#data['Age'] = (pd.to_datetime('today') - data['Dt_Customer']).dt.days // 365
data['Total_Spending'] = data['MntWines'] + data['MntFruits'] + data['MntMeatProducts'] + data['MntFishProducts'] + data['MntSweetProducts'] + data['MntGoldProds']
```

<div style="text-align:left; color:Blue; font-size:25px;"><b>STEP 3(b)</b></div>
Derive the total purchases from the number of transactions across the three channels.

```python
#Creating a new column 'Total_Purchases' to represent the total number of purchases made by each customer
data['Total_Purchases'] = data['NumWebPurchases'] + data['NumCatalogPurchases'] + data['NumStorePurchases'] + data['NumDealsPurchases']
data.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Generating box plot and histogram to visualize the distribution of Income column before outlier treatment
plt.figure(figsize=(12, 5))

# Box plot
plt.subplot(1, 2, 1)
sns.boxplot(y=data['Income'])
plt.title('Income Box Plot')

# Histogram with KDE
plt.subplot(1, 2, 2)
sns.histplot(data['Income'], bins=30, kde=True)
plt.axvline(data['Income'].median(), color='red', linestyle='--', label='Median')
plt.axvline(data['Income'].mean(), color='green', linestyle='--', label='Mean')
plt.title('Income Distribution')
plt.legend()

plt.tight_layout()
plt.show()
```

<div style="text-align:left; color:yellow; font-size:20px;">Observation:<br></div>
<div style="text-align:left; color:white; font-size:15px;">
<ol>
<li>Income is right-skewed as mean income lies to the right of the median, indicates a small number of high-income customers pulling the average upward</li>
<li>Box plot shows several points far above the upper whisker confirming the presence of extreme high-income outliers</li>
<li>Majority of customers fall within a moderate income range</li>
<li>Lower tail is relatively short, Few very low-income values compared to high-income extremes</li></ol></div>

```python
#Outlier Treatment for Income column using Winsorization
from scipy.stats.mstats import winsorize
data['Income_raw'] = data['Income'].copy()
data['Income_winsorized'] = winsorize(
    data['Income_raw'],
    limits=[0.01, 0.01]   # 1% lower, 1% upper
)
```

```python
#Visualizing the effect of Winsorization after outlier treatment
plt.figure(figsize=(12, 5))

# Original Income
plt.subplot(2, 2, 1)
sns.boxplot(y=data['Income'])
plt.title('Original Income')

# Winsorized Income
plt.subplot(2, 2, 2)
sns.boxplot(y=data['Income_winsorized'])
plt.title('Winsorized Income')

# Histogram of Income
plt.subplot(2, 2, 3)
sns.histplot(data['Income'], bins=25, kde=True)
plt.axvline(data['Income'].median(), color='red', linestyle='--', label='Median')
plt.axvline(data['Income'].mean(), color='green', linestyle='--', label='Mean')
plt.title('Income Distribution')
plt.legend()

# Histogram of Winsorized Income
plt.subplot(2, 2, 4)
sns.histplot(data['Income_winsorized'], bins=25, kde=True)
plt.axvline(data['Income_winsorized'].median(), color='red', linestyle='--', label='Median')
plt.axvline(data['Income_winsorized'].mean(), color='green', linestyle='--', label='Mean')
plt.title('Winsorized Income Distribution')
plt.legend()

plt.tight_layout()
plt.show()
```

<div style="text-align:left; color:yellow; font-size:20px;">Observation after outlier treatment:<br></div>
<div style="text-align:left; color:white; font-size:15px;">
<ol>
<li>Extreme high-income outliers have been effectively capped as the very large income values seen in the original box plot are no longer present</li>

<div style="text-align:left; color:Blue; font-size:25px;"><b>STEP 7(a)</b></div>
considering with 95 % conf and p_value = 0.05<br>
Older individuals may not possess the same level of technological proficiency and may, therefore, lean toward traditional in-store shopping preferences.

```python
older_customers = data[data['Age'] > 60]
```

```python
from scipy.stats import ttest_ind

t_score, p_value = ttest_ind(
    older_customers['NumStorePurchases'],
    older_customers['NumWebPurchases'],
    equal_var=False  # Welch’s t-test (safer)
)

t_score, p_value
```

```python
if p_value < 0.05:
  print("reject null hypothesis")
else:
  print("accept null hypothesis")
```

<div style="text-align:left; color:yellow; font-size:20px;">Observation:<br></div>
<div style="text-align:left; color:white; font-size:15px;">
<b>Reject null hypothesis:</b> There is a statistically significant difference between store and web purchases among older customers.
The t_score > 12: meaning, mean(NumStorePurchases) > mean(NumWebPurchases)
Since the mean number of store purchases is higher than web purchases, older customers show a preference for in-store shopping.</div>

<div style="text-align:left; color:Blue; font-size:25px;"><b>STEP 7(b)</b></div>
Customers with children likely experience time constraints, making online shopping a more convenient option.

```python
Cust_child = data[data['Total_Children'] > 1]
t_stat, p_value = ttest_ind(
    Cust_child['NumStorePurchases'],
    Cust_child['NumWebPurchases'],
)

t_stat, p_value
```

```python
if p_value < 0.05:
  print("reject null hypothesis")
else:
  print("accept null hypothesis")
```

<div style="text-align:left; color:yellow; font-size:20px;">Observation:<br></div>
<div style="text-align:left; color:white; font-size:15px;">
<b>Reject null hypothesis Observation:</b> There is a statistically significant difference between the mean number of store purchases and web purchases for customers having more than one child. <div style="text-align:left; color:Blue; font-size:25px;"><b>STEP 7(c)</b></div>
Sales at physical stores may face the risk of cannibalization by alternative distribution channels.

```python
data['other_purchases'] = data['NumCatalogPurchases'] + data['NumWebPurchases']
t_score, p_value = ttest_ind(
    data['NumStorePurchases'],
    data['other_purchases'],
    #equal_var=False
)
t_score, p_value
```

```python
if p_value < 0.05:
  print("reject null hypothesis")
else:
  print("accept null hypothesis")
```

<div style="text-align:left; color:white; font-size:15px;">
<b>Reject null hypothesis Observation:</b> There is a statistically significant difference between the average number of store purchases and other purchases. Customers, on average, make significantly more purchases via online and catalog channels combined than through in-store purchases which is derived from t_score value.</div>
